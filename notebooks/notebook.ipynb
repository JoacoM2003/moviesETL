```python
import requests
import pandas as pd
```

## Obtención de datos

```python
# Extraer varias páginas
all_shows = []
for page in range(5):
    url = f"https://api.tvmaze.com/shows?page={page}"
    resp = requests.get(url)
    all_shows.extend(resp.json())

# Convertir a DataFrame y explorar
df = pd.DataFrame(all_shows)
df.info()
df.head()
```

## Transform

### Qué columnas voy a usar

```python
# Elegir columnas utiles
df_clean = df[[
    'id', 'name', 'genres', 'status', 'language', 'premiered', 'network'
]]
# Extraer rating promedio
df_clean['rating'] = df['rating'].apply(lambda x: x['average'] if x else None)
df_clean
```

### Manejo de nulos

```python
# Miro los nulos y luego veo en cada columna los que hay
df_clean.isnull().sum()
```

```python
# Premiered hay un nulo, no molesta por lo que lo dejo como nulo.
# Ratings tiene muchos nulos, puedo sacar la media o dejarlos como nulos, no molestan.
# Network tiene nulos, unknown
df_clean["network"] = df_clean["network"].fillna("Unknown")

df_clean.isnull().sum()
```

### Manejo de duplicados

```python
# Chequeo que no haya duplicados en los id
df_clean.duplicated(subset=['id']).sum()
df_clean
```

### Normalización

```python
# Ver los tipos de datos de las columnas
df_clean.dtypes
# premeried debería ser datetime
df_clean['premiered'] = pd.to_datetime(df_clean['premiered'], errors='coerce')
# rating debería ser float
df_clean['rating'] = pd.to_numeric(df_clean['rating'], errors='coerce')
df_clean.dtypes
```

```python
# Paso a Mayusculas
df_clean['status'] = df_clean['status'].str.capitalize()
df_clean['language'] = df_clean['language'].str.capitalize()
```

### Acomodar la columna Network

```python
# Extraer 'name' de network, manejar nulos
df_clean['network'] = df_clean['network'].apply(lambda x: x['name'] if isinstance(x, dict) and x.get('name') else "Unknown")
df_clean
```

## Enriquecimiento

```python
# Separo los generos en distintas filas
df_exploded = df_clean.explode('genres')
df_exploded
```

## Load
